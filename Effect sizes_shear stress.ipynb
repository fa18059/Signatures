```python
#import libraries
import pandas as pd
import matplotlib.pyplot as plt
```

```python
#get data from excel workbook
file_loc = "Effect sizes_data frame.xlsx"
df = pd.read_excel(file_loc, 'Shear stress', index_col=None) #call specific sheet as a dataframe

#create figure and set size
fig = plt.figure(figsize=(10,7))

#creating dictionary to match impacts with colours
colours = {'Impact 1':'#fad3ff', 'Impact 2':'#c099d7', 'Impact 3':'#835dad'}

#group x and y values by impacts, in order to plot
for k,g in df.groupby('Impact'):
    plt.scatter(g['lnRR'],g['Paper'],label=k,c=g['Impact'].map(colours), s = 60) #calling specific columns 

#format figure
plt.title('Shear stress variables', fontsize=16) #chart title
plt.axvline(x = 0, color = 'black', linewidth=0.5) #vertical line at x=0
plt.legend(loc='upper right', fontsize=12) #legend location and formatting
fig.set_facecolor("w") #white background

#format axes
plt.yticks(fontsize=14)
plt.xticks(range(-1,2), fontsize=14) #set x axis range to be even on either side of x=0
plt.xlabel('lnRR', fontsize=14) #x axis label
plt.ylabel('Paper', fontsize=14) #y axis label

#indicate freshwater vs terrestrial via y axis label colours
for j in range(0,5):
    plt.gca().get_yticklabels()[j].set_color('mediumblue') #for freshwater

#save plot as png
plt.savefig('Shear stress_bypaper_col.png', bbox_inches='tight', dpi=300) #bbox tight prevents cutting off of labels
plt.show() #show plot below
```
