```python
#import libraries
import pandas as pd
import matplotlib.pyplot as plt
import sys
!{sys.executable} -m pip install pysankey #had to download this package because never used before 
                                            #can just use 'pip install pysankey' if have used it previously
from pySankey.sankey import sankey
```

```python
#get data from csv
data = pd.read_csv('T_sankey.csv')

#create dictionary for colours
colors = {
    "Malacostraca": "blue",
    "Insecta": "#d95f02",
    "Reptilia": "#1b9e77",
    "Clitellata": "#7570b3",
    "Aves": "#e7298a",
    "Arachnida": "#66a61e",
    "Mammalia": "#e6ab02",
    "Bioprotection": "black",
    "Bioerosion": "black",
    "Bioconstruction": "black",
    "Bioturbation": "black"}

#plot sankey using column headings from csv
sankey(
    left=data["Class"], right=data["Biogeo"], 
    leftWeight= data["Weight"], rightWeight=data["Weight"], 
    aspect=20, fontsize=12, colorDict=colors)

fig = plt.gcf() #get current figure
fig.set_size_inches(6, 6) #set size
fig.set_facecolor("w") #white background

#save plot as png
fig.savefig("T_sankey.png", bbox_inches="tight", dpi=300) #bbox tight prevents cutting off of labels
```
