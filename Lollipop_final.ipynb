```python
#import libraries
import pandas as pd
import matplotlib.pyplot as plt
```

```python
#get data from csv
dataraw = pd.read_csv("Lollipop.csv") #read whole csv
data = pd.read_csv("Lollipop.csv", index_col ="Taxonomic class") #read whole csv but using a specified index column

#rename data columns for easier use in later code
species = data["Unique species"]
papers = data["Papers"] 
```

```python
dataraw = dataraw.iloc[::-1] #reverse the order of the data so the first row of the csv is plotted at the top of the plot

#create a list for y axis labels, make sure the order matches the order of the csv values
animals = ['Aves','Aves','Clitellata','Clitellata','Insecta','Insecta','Malacostraca','Malacostraca','Mammalia','Mammalia','Arachnida','Reptilia','Actinopterygii','Amphibia','Bivalvia']
label_anim = list(animals)
my_range=range(1,len(data.index)+1) #specifying the length of the lollipop lines, based on the data points

#create figure and set size
fig = plt.figure(figsize = (10,7))
ax = plt.axes(frameon=True) #gives frame/border
ax.grid(False) #removes gridlines

#plotting the data
plt.hlines(y=my_range, xmin=species, xmax=papers, color='#8f9c9a', lw=2) #lollipop lines
plt.scatter(species, my_range, color='orange', s=75, label='No. of unique species', zorder=3) #species points
plt.scatter(papers, my_range, color='dimgrey', s=75, label='No. of papers', zorder=3) #papers points

#format figure
plt.legend(ncol=1, loc='upper right', frameon=False, fontsize=14) #legend location and formatting
plt.yticks(my_range, label_anim, fontsize=14) #y axis labels
plt.xticks(fontsize=14) #x axis labels
ax.set_xlim(0) #make sure y axis crosses at x=0
plt.xlabel('Frequency', fontsize=14) #x axis title
plt.ylabel('Taxonomic class', fontsize=14) #y axis title
fig.set_facecolor("w") #white background

#indicate freshwater vs terrestrial via y axis label colours
ax.get_yticklabels()[0].set_color("saddlebrown") #for terrestrial
ax.get_yticklabels()[1].set_color("mediumblue") #for freshwater
ax.get_yticklabels()[2].set_color("saddlebrown")
ax.get_yticklabels()[3].set_color("mediumblue")
ax.get_yticklabels()[4].set_color("saddlebrown")
ax.get_yticklabels()[5].set_color("mediumblue")
ax.get_yticklabels()[6].set_color("saddlebrown")
ax.get_yticklabels()[7].set_color("mediumblue")
ax.get_yticklabels()[8].set_color("saddlebrown")
ax.get_yticklabels()[9].set_color("mediumblue")
ax.get_yticklabels()[10].set_color("saddlebrown")
ax.get_yticklabels()[11].set_color("saddlebrown")
ax.get_yticklabels()[12].set_color("mediumblue")
ax.get_yticklabels()[13].set_color("mediumblue")
ax.get_yticklabels()[14].set_color("mediumblue")
ax.invert_yaxis() #to make sure the points and the labels match up correctly

#adding to legend
plt.text(54.8, 3.8, 'Freshwater', fontsize = 14, color='mediumblue') 
plt.text(54.8, 3, 'Terrestrial', fontsize = 14, color='saddlebrown')
rect = plt.Rectangle((49.7, 0.6), 27, 3.5, edgecolor='grey', facecolor='none') #create frame around legend
ax.add_patch(rect)

#save plot as png
plt.savefig('Lollipop_final2.png', bbox_inches='tight', dpi=300) #bbox tight prevents cutting off of labels
plt.show() #show plot below
```
