```python
#import libraries
import numpy as np
import matplotlib.pyplot as plt

#get data from csv
all_data = np.genfromtxt('Violin_plots.csv', delimiter=',')

#create figure and set size
fig = plt.figure(figsize=(7,10))

#plot horizontal violins for each variable. must state exact number of rows with data and specific column in csv
sediment_transport = plt.violinplot(all_data[1:29, 0:1], positions=[1], showextrema=False, showmedians=True,vert=False)
erosion = plt.violinplot(all_data[1:35, 5:6], positions=[2], showextrema=False, showmedians=True,vert=False)
shear_stress = plt.violinplot(all_data[1:16, 2:3], positions=[3], showextrema=False, showmedians=True,vert=False)
porosity = plt.violinplot(all_data[1:17, 3:4], positions=[4], showextrema=False, showmedians=True,vert=False)
infiltration = plt.violinplot(all_data[1:27, 4:5], positions=[5], showextrema=False, showmedians=True,vert=False)
sediment_compaction = plt.violinplot(all_data[1:38, 1:2], positions=[6], showextrema=False, showmedians=True,vert=False)
clay = plt.violinplot(all_data[1:26, 6:7], positions=[7], showextrema=False, showmedians=True,vert=False)

#set violin colours for each variable. alpha refers to transparency
for pc in sediment_transport['bodies']:
    pc.set_facecolor('#a6bddb')
    pc.set_edgecolor('#a6bddb')
    pc.set_alpha(0.7)
    
for pc in erosion['bodies']:
    pc.set_facecolor('#a6bddb')
    pc.set_edgecolor('#a6bddb')
    pc.set_alpha(0.7)
    
for pc in shear_stress['bodies']:
    pc.set_facecolor('#a6bddb')
    pc.set_edgecolor('#a6bddb')
    pc.set_alpha(0.7)
    
for pc in porosity['bodies']:
    pc.set_facecolor('#a6bddb')
    pc.set_edgecolor('#a6bddb')
    pc.set_alpha(0.7)
    
for pc in infiltration['bodies']:
    pc.set_facecolor('#a6bddb')
    pc.set_edgecolor('#a6bddb')
    pc.set_alpha(0.7)
    
for pc in sediment_compaction['bodies']:
    pc.set_facecolor('#a6bddb')
    pc.set_edgecolor('#a6bddb')
    pc.set_alpha(0.7)
    
for pc in clay['bodies']:
    pc.set_facecolor('#a6bddb')
    pc.set_edgecolor('#a6bddb')
    pc.set_alpha(0.7)
    
#set solour of median lines for each variable
sediment_transport['cmedians'].set_color('k')
erosion['cmedians'].set_color('k')
shear_stress['cmedians'].set_color('k')
porosity['cmedians'].set_color('k')
infiltration['cmedians'].set_color('k')
sediment_compaction['cmedians'].set_color('k')
clay['cmedians'].set_color('k')

#organise axes formatting
ax = plt.gca() #get current axes
labels = ax.get_yticklabels() #naming y axis labels below
labels = ['Sediment transport','Erosion' , 'Shear stress', 'Soil porosity', 'Infiltration', 'Sediment compaction', 'Clay']
ax.set_yticks([1,2,3,4,5,6,7]) #choose positions of the y axis labels
ax.set_yticklabels(labels, fontsize='14') #set y axis labels
plt.xticks(range(-6,7), fontsize='14') #x axis labels
ax.set_ylim(0.5, len(labels) + 0.5) #set y axis limits
ax.set_ylabel('Variable groups', fontsize='16') #y axis title
ax.set_xlabel('lnRR', fontsize='16') #x axis title
plt.axvline(x = 0, color = 'black', linewidth=0.5, alpha=0.5) #line at x=0
fig.set_facecolor("w") #white background

plt.axhline(y=10, linestyle='-',color='k',label='Median', linewidth = 2) #median line in legend
plt.legend(fontsize = 14)

#save plot as png
plt.savefig('Violin_effectvariables_vert.png', bbox_inches='tight', dpi=300) #bbox tight prevents cutting off of labels
plt.show() #show plot below
```
